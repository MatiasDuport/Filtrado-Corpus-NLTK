```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag


import re

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('cess_esp')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')


lemmatizer = WordNetLemmatizer()
```

Sin filtrar

```python
print(nltk.data.path)
```

```python
corpus = nltk.corpus.cess_esp.sents()
corpus_flatten = [palabra for oracion in corpus for palabra in oracion]
print(corpus_flatten)
```

```python
len(corpus_flatten)
```

Filtrado

```python
stop_words = set(stopwords.words('spanish'))
print(stop_words)
```

```python

# Función para obtener la etiqueta POS de WordNet
def get_wordnet_pos(word):
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
corpus_filtrado = [
    lemmatizer.lemmatize(re.sub(r'[^\w\s]', '', palabra.lower()), get_wordnet_pos(palabra))
    for palabra in corpus_flatten
    if palabra.lower() not in stop_words  # Quitar stop words
    and re.sub(r'[^\w\s]', '', palabra.lower()) != ''  # Elimina palabras vacías tras quitar puntuación
]

print(corpus_filtrado)
```
